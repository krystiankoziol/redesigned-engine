```python
def Read_Epoch(file_name, n):
  with open(file_name, 'r') as f:
    data = np.loadtxt(f, skiprows= 1, usecols=(n, n+1))
  return data    
```

```python
n  = 1
filename = "pkt_cosinusy.txt"
thres = 1.5
data = Read_Epoch(filename, 0)
print(data)
```

```python
for i in data:
  for id in data:
    dx = i[0] - id[0]
    dy = i[1] - id[1]
    f = np.arctan(abs(dy/dx))
    if dx and dy > 0:
      Az = f
    elif dx > 0 and dy < 0:
      Az = 400 - f
    elif dx < 0 and dy < 0:
      Az = 200 + f
    else:
      Az = 200 - f
    print(Az)
```
